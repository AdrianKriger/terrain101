### hw01_a:Interpolation

#### Exploratory analysis of .xyz datasets - [these specifically](https://3d.bk.tudelft.nl/courses/backup/geo1015/2018/hw/01/) - and visualization of results.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image, HTML, display
```

#### samples.xyz

```python
f = 'C:/{path}/samples.xyz'
df = pd.read_csv(f, delimiter=' ', header='infer')
```

```python
df.head(2)
```

```python
df.describe()
```

```python
print('diff x:', df.x.min() - df.x.max())
print('diff y:', df.y.min() - df.y.max())
print('diff z:', df.z.min() - df.z.max())
```

```python
sns.displot(df.z)
```

```python
extent = x_min, x_max, y_min, y_max = [df.x.min()-25, df.x.max()+25,
                                       df.y.min()-25, df.y.max()+25]
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(df.x, df.y, c=df.z, s=4)
ax.set_aspect(1)
ax.set_xlim(*extent[:2])
ax.set_ylim(*extent[2:])
ax.set_xlabel('Easting [m]')
ax.set_ylabel('Northing [m]')
ax.set_title('Samples.xyz')
ax.grid(c='k', alpha=0.2)

plt.show()
```
